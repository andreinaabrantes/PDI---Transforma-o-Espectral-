#Visualização com Earthpy

```python
bandas_ep = []

for i in sorted(glob.glob('/content/Bandas/*[1-8].tif')):
  bandas_ep.append(i)
```

```python
stack2, metadata = es.stack(bandas_ep)
```

```python
#composição de cor real 
ep.plot_rgb(stack2, rgb=((2,1,0)), stretch = True, str_clip=2)
```

```python
#composição ressaltando a vegetação
ep.plot_rgb(stack2, rgb=((2,3,0)), stretch = True, str_clip=2)
```
